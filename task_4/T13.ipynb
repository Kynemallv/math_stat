```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

n = 139
m = 1000
```

```python
c1 = float(st.f.ppf(0.025, n - 1, m - 1))
c2 = float(st.f.ppf(0.975, n - 1, m - 1))
print(f"{c1 = }\n{c2 = }")
```

```python
x = np.linspace(-3, 3, 1000)

fig, ax = plt.subplots(1, 1)
ax.plot(x, st.f.sf(c2 * x, n - 1, m - 1) + st.f.cdf(c1 * x, n - 1, m - 1))
```
